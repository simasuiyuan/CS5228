```python
from math import sin, cos, sqrt, atan2, radians
def compute_dist(lat1:float, lng1:float, lat2:float, lng2:float)->float:
    # approximate radius of earth in km
    R = 6373.0
    lat1 = radians(lat1)
    lng1 = radians(lng1)
    lat2 = radians(lat2)
    lng2 = radians(lng2)
    dlng = lng2 - lng1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlng / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c
    
def distance_to_closest_centres(row):
    shortest_distances = {
        'CR_dist': np.inf,
        'CR': None,
        'IEBP_dist': np.inf,
        'IEBP': None,
        'BN_dist': np.inf,
        'BN': None,
        'IHL_dist': np.inf,
        'IHL': None,
    }
    for type in sg_commerical_centres_df["type"].unique():
        available_centers = sg_commerical_centres_df.loc[sg_commerical_centres_df["type"]==type, ['lat','lng','name']]
        available_centers['dist'] = available_centers.apply(lambda available_center: compute_dist(row['lat'], row['lng'], available_center['lat'], available_center['lng']), axis=1)
        shortest_distances[f"{type}_dist"] =  available_centers["dist"].min()
        shortest_distances[type] =  available_centers.loc[available_centers["dist"]==available_centers["dist"].min(),'name'].values[0]
    return shortest_distances
```

```python
# print(train_df_clean.shape)
unique_location = train_df_clean[['lat', 'lng']].drop_duplicates().reset_index(drop=True)
closest_commercial_centers = unique_location.merge(pd.DataFrame(list(unique_location.apply(distance_to_closest_centres, axis=1))), left_index=True, right_index=True)
closest_commercial_centers.head()
```

```python
train_df_clean = train_df_clean.merge(closest_commercial_centers, left_on=['lat','lng'], right_on=['lat','lng'])
print(train_df_clean.shape)
train_df_clean.head()
```

```python
from math import sin, cos, sqrt, atan2, radians
def compute_dist(lat1:float, lng1:float, lat2:float, lng2:float)->float:
    # approximate radius of earth in km
    R = 6373.0
    lat1 = radians(lat1)
    lng1 = radians(lng1)
    lat2 = radians(lat2)
    lng2 = radians(lng2)
    dlng = lng2 - lng1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlng / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c
    
def distance_to_closest_mrt_station(row, radius=0.5):
    shortest_distances = {
        'mrt_dist': np.inf,
        'mrt_station': None,
        'number_of_close_stations':0,
        'code':None,
        'line':None
    }

    available_stations = sg_mrt_stations_df.loc[:, ['lat','lng','name','code','line']]
    available_stations['dist'] = available_stations.apply(lambda available_station: compute_dist(row['lat'], row['lng'], available_station['lat'], available_station['lng']), axis=1)
    shortest_distances["mrt_dist"] =  available_stations["dist"].min()
    shortest_distances['mrt_station'] = available_stations.loc[available_stations["dist"]==available_stations["dist"].min(),'name'].values[0]
    shortest_distances['number_of_close_stations'] = available_stations.loc[available_stations["dist"] <= radius].shape[0]
    shortest_distances['code'] = available_stations.loc[available_stations["dist"]==available_stations["dist"].min(),'code'].values[0]
    shortest_distances['line'] =  available_stations.loc[available_stations["dist"]==available_stations["dist"].min(),'line'].values[0]
    return shortest_distances
```

```python
unique_location = train_df_clean[['lat', 'lng']].drop_duplicates().reset_index(drop=True)
closest_mrt_station = unique_location.merge(pd.DataFrame(list(unique_location.apply(distance_to_closest_mrt_station, axis=1))), left_index=True, right_index=True)
closest_mrt_station.head()
```

```python
train_df_clean = train_df_clean.merge(closest_mrt_station, left_on=['lat','lng'], right_on=['lat','lng'])
print(train_df_clean.shape)
train_df_clean.head()
```
